```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('output-single/link_kpis.csv')
df.head()
```

```python
linkData = df[df['start_stop'] == 'PF_A15_C']
linkData
```

```python
# plotting the mean headway
fig, ax = plt.subplots(figsize=(20, 5))

linkData.plot(x='datetime', y='mean_headway [mins]', ax=ax, label='Mean Headway', color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Headway [min]')
ax.set_title('Mean Headway for Link PF_A15_C')

plt.legend()
plt.show()
```

```python
#plotting the mean speed
fig, ax = plt.subplots(figsize=(20, 5))

linkData.plot(x='datetime', y='mean_speed [km/h]', ax=ax, label='Mean Speed', color='red')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Speed [km/h]')
ax.set_title('Mean Speed for Link PF_A15_C')

plt.legend()
plt.show()
```

```python
# plotting the mean time 
fig, ax = plt.subplots(figsize=(20, 5))

linkData.plot(x='datetime', y='mean_time [mins]', ax=ax, label='Mean Time', color='green')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Time [min]')
ax.set_title('Mean Time for Link PF_A15_C')

plt.legend()
plt.show()
```
