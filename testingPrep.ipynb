```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('output-single/link_kpis_total.csv')
df['datetime'] = pd.to_datetime(df['datetime'])
df.sort_values('datetime', inplace=True)
df.set_index('datetime', inplace=True)
df.head()
```

```python
# removing outliers
df = df[df['mean_time [mins]'] < 100]
df = df[df['mean_headway [mins]'] < 100]
df = df[df['mean_speed [km/h]'] < 120]
```

```python
linkData = df[df['start_stop'] == 'PF_A13_C']
linkData
```

```python
# plotting the mean headway
fig, ax = plt.subplots(figsize=(20, 5))
linkData['mean_headway [mins]'].plot(ax=ax, label='Mean Headway')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Headway [min]')
ax.set_title('Mean Headway for Link PF_A15_C')

plt.legend()
plt.show()
```

```python
#plotting the mean speed
fig, ax = plt.subplots(figsize=(20, 5))

linkData['mean_speed [km/h]'].plot(ax=ax, label='Mean Headway')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Speed [km/h]')
ax.set_title('Mean Speed for Link PF_A15_C')

plt.legend()
plt.show()
```

```python
# plotting the mean time 
fig, ax = plt.subplots(figsize=(20, 5))

linkData['mean_time [mins]'].plot(ax=ax, label='Mean Headway')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Time [min]')
ax.set_title('Mean Time for Link PF_A15_C')

plt.legend()
plt.show()
```

```python
# Reset index to have the datetime as a column
df_reset = df.reset_index()

# Group by start_stop, end_stop, and resample on datetime
grouped_resampled = df_reset.groupby(
    ['start_stop', 'end_stop', pd.Grouper(key='datetime', freq='15T')]
).apply(
    lambda g: pd.Series({
        'weighted_avg_headway': np.average(g['mean_headway [mins]'], weights=g['no_of_trips']),
        'weighted_avg_speed':   np.average(g['mean_speed [km/h]'],   weights=g['no_of_trips']),
        'weighted_avg_time':    np.average(g['mean_time [mins]'],    weights=g['no_of_trips']),
        'no_of_trips':          g['no_of_trips'].sum()
    })
).reset_index()

grouped_resampled.head()





```

```python
# plotting the mean headway
fig, ax = plt.subplots(figsize=(20, 5))
grouped_resampled[grouped_resampled['start_stop'] == 'PF_A15_C']['weighted_avg_headway'].plot(ax=ax, label='Mean Headway')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Headway [min]')
ax.set_title('Mean Headway for Link PF_A15_C')

plt.legend()
plt.show()
```

```python
#plotting the mean speed
fig, ax = plt.subplots(figsize=(20, 5))

grouped_resampled[grouped_resampled['start_stop'] == 'PF_A15_C']['weighted_avg_speed'].plot(ax=ax, label='Mean Headway')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Speed [km/h]')
ax.set_title('Mean Speed for Link PF_A15_C')

plt.legend()
plt.show()
```

```python
# plotting the mean time
fig, ax = plt.subplots(figsize=(20, 5))

grouped_resampled[grouped_resampled['start_stop'] == 'PF_A15_C']['weighted_avg_time'].plot(ax=ax, label='Mean Headway')
ax.set_xlabel('Time')
ax.set_ylabel('Mean Time [min]')
ax.set_title('Mean Time for Link PF_A15_C')

plt.legend()
plt.show()
```
